Imagina que un grupo de amigos quiere saber más sobre su cosecha de manzanas.
Cosechan 10 manzanas y quieren analizar su peso para ver cómo es su producción este año.
Queremos calcular:
Media (promedio), Mediana (peso central) Moda (peso más común, Desviación estandar (lo que varían los pesos)

```python
import numpy as np
import statistics as stats

manzanas = [150, 180, 200, 170, 150, 190, 200, 210, 150, 180]

#Estadísticas descriptivas
media = np.mean(manzanas) #promedio
mediana = np.median(manzanas) #valor central
moda = stats.mode(manzanas) #peso más común
desviacion = np.std(manzanas) #lo que varían los pesos

#Resultados
print(f"Análisis de la cosecha de manzanas 🍎")
print(f"Media: {media}g")
print(f"Mediana: {mediana}g")
print(f"Moda: {moda}g")
print(f"Desviación estándar: {desviacion:.2f}g")
```
