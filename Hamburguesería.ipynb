```python
inventario = 100
print(f"El inventario contiene{inventario} hamburguesas")
while inventario > 0:
    if inventario <= 10:
      print("ADVERTENCIA:EL INVENTARIO ESTÁ CASI VACÍO")
    num_hamburguesas = int(input("¿Cuántas haburguesas desea comprar?"))
    if num_hamburguesas > inventario:
      print("No hay suficientes hamburguesas, solo quedan",inventario, "disponibles")
    else:
      inventario -= num_hamburguesas
      print (f"Haz pedido", num_hamburguesas, "Solo nos quedan", inventario, "hamburguesas")
```

```python
# Inventario inicial (en unidades)
manzanas = 30
platanos = 20
naranjas = 25

LIMITE_AVISO = 5  # cuando queden 5 o menos, avisamos

while True:
    print("\n🍎 INVENTARIO ACTUAL")
    print(f"Manzanas: {manzanas}")
    print(f"Plátanos: {platanos}")
    print(f"Naranjas: {naranjas}")

    # Avisos de stock bajo
    if manzanas <= LIMITE_AVISO:
        print("⚠️ Aviso: quedan pocas MANZANAS")
    if platanos <= LIMITE_AVISO:
        print("⚠️ Aviso: quedan pocos PLÁTANOS")
    if naranjas <= LIMITE_AVISO:
        print("⚠️ Aviso: quedan pocas NARANJAS")

    # Si todo está a 0, cerramos
    if manzanas == 0 and platanos == 0 and naranjas == 0:
        print("\n🚫 No queda nada. Cerrando frutería.")
        break

    fruta = input("\n¿Qué fruta quieres comprar? (manzanas/platanos/naranjas o salir): ").lower()

    if fruta == "salir":
        print("👋 Hasta luego")
        break

    cantidad = int(input("¿Cuántas unidades quieres?: "))

    if fruta == "manzanas":
        if cantidad > manzanas:
            print(f"❌ No hay suficientes. Solo quedan {manzanas} manzanas.")
        else:
            manzanas -= cantidad
            print(f"✅ Has comprado {cantidad} manzanas. Quedan {manzanas}.")

    elif fruta == "platanos":
        if cantidad > platanos:
            print(f"❌ No hay suficientes. Solo quedan {platanos} plátanos.")
        else:
            platanos -= cantidad
            print(f"✅ Has comprado {cantidad} plátanos. Quedan {platanos}.")

    elif fruta == "naranjas":
        if cantidad > naranjas:
            print(f"❌ No hay suficientes. Solo quedan {naranjas} naranjas.")
        else:
            naranjas -= cantidad
            print(f"✅ Has comprado {cantidad} naranjas. Quedan {naranjas}.")

    else:
        print("❌ Esa fruta no existe. Escribe manzanas, platanos, naranjas o salir.")
```
